# Logistic Regression of Mine

```python
import pandas as pd
import numpy as np 

class Logistic_Regression:
    def __init__(self,learning_rate=0.01,num_iterations=100):
        """
        learning_rate: 控制梯度下降幅度的超参数
        num_iterations: 梯度迭代次数 —— 不合理会导致欠拟合或过拟合
        weights: 表示每个特征维度的权重
        bias: 偏置 即 常数项 没有任何输入时的输出值
        """
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None
        

    def Sigmoid(self,z):
        """
        Sigmoid: 类S型曲线函数
        """
        return 1/(1 + np.exp(-z))
    
    def fit(self,X,y):
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        # 梯度下降法
        for iteration in range(self.num_iterations) : 
            y_pred = self.Sigmoid(np.dot(X,self.weights) + self.bias)

            # 梯度
            dw = (1/X.shape[0]) * np.dot(X.T,(y_pred-y))
            db = (1/X.shape[0]) * np.sum(y_pred-y)
            
            # 更新权重
            self.weights -= self.learning_rate  * dw
            self.bias  -= self.learning_rate  * db  

    def predict(self,X):
        y_pred = self.Sigmoid(np.dot(X,self.weights)+ self.bias)
        return (y_pred>0.5).astype(int)
```

##### Draw the Sigmoid Function

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_sigmoid():
    x = np.linspace(-10, 10, 100000)
    y = 1/(1 + np.exp(-x))

    plt.plot(x, y, label='Sigmoid Function')
    plt.xlabel('x')
    plt.ylabel('y = 1/(1+exp(-x))')
    plt.box(False)

    plt.ylim([-0.2, 1.2])
    plt.xlim([-15,15])
    plt.legend()

    plt.axhline(0,color = 'red',linestyle='--')
    plt.axhline(1,color = 'red',linestyle='--')
    plt.axhline(0.5, color='black', linestyle='-')
    plt.axhline(-0.2, color='black', linestyle='-', linewidth=1)
    plt.axvline(-15, color='black', linestyle='-')
    plt.axvline(0, color='black', linestyle='-', linewidth=1)

    plt.show()
draw_sigmoid()
```
